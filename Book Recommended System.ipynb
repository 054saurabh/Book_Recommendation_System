```python
x=df.groupby('User-ID').count()['Book-Rating']>200
padhe_users=x[x].index
```

```python
filter_rating=df[df['User-ID'].isin(padhe_users)]
```

```python
y=filter_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_book=y[y].index
```

```python
final_ratings=df[df['Book-Title'].isin(famous_book)]
```

```python
final_ratings.shape
```

```python
final_ratings.drop_duplicates()
```

```python
pre_df=final_ratings.pivot_table(index='Book-Title', columns='User-ID',values='Book-Rating')
```

```python
pre_df
```

```python
pre_df.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cs=cosine_similarity(pre_df)
```

```python
cs.shape
```

```python
cs
```

```python
def recomend(book_name):
    index_num=np.where(pre_df.index==book_name)[0][0]
    similar=sorted(list(enumerate(cs[index_num])),key=lambda p:p[1],reverse=True)[1:6]
    for i in similar:
        print(pre_df.index[i[0]])
```

```python
recomend('The Catcher in the Rye')
```
